### Custom Functions

```python
# using def
def sqrt(x):
    return x**0.5

print(sqrt(4))

def pythagoras(a, b):
    return (a**2 + b**2)**0.5

print(pythagoras(3, 4))

# using lambda
sqrt_lam = lambda x: x**0.5
print(sqrt_lam(4))

pythagoras_lam = lambda a, b: (a**2 + b**2)**0.5
print(pythagoras_lam(3, 4))
```

```python
# exercise
import numpy as np

def dollars_to_idr(dollar):
    return dollar * 14383

print(dollars_to_idr(1000))

def tri_area(a, t):
    return a * t / 2
print(tri_area(2, 10))

yen_to_idr = lambda yen: yen * 109.35
print(yen_to_idr(1000))

volume_cone = lambda r, t: 1/3 * (np.pi * r**2) * t
print(volume_cone(10, 5))
```

### Merging Dataframe

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
# create 2 dataframes
titanic_c = df.iloc[11:16, [0, 2, 3, 4, 5]]
titanic_d = df.iloc[13:18, [0, 7, 8, 9, 10, 11]]

titanic_c.head()
```

```python
titanic_d.head()
```

```python
# merging using .join() method
titanic_e = titanic_c.add_suffix("_e") # to differentiate
titanic_e.join(titanic_d, how="outer")
```

```python
# merging using .merge() method
titanic_c.merge(titanic_d, how="inner")
```

```python
# merging using pd.concat() method
pd.concat([titanic_c, titanic_d], axis=1)
```
